```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\BHARGAV\Desktop\insurance-cost-prediction\Datasets\insurance.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import joblib
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
cat_cols
```

```python
df_encoded = df.drop(columns=cat_cols)
```

```python
for col in cat_cols:
    ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    
    # Fit and transform
    encoded_data = ohe.fit_transform(df[[col]])
    
    # Make sure the columns match exactly the data shape
    col_names = ohe.get_feature_names_out([col])
    
    # Check shapes before DataFrame creation
    # print(f"Encoding '{col}': data shape {encoded_data.shape}, columns {len(col_names)}")
    
    encoded_df = pd.DataFrame(encoded_data, columns=col_names)
    
    df_encoded = pd.concat([df_encoded, encoded_df], axis=1)
    
    joblib.dump(ohe, f'ohe_{col}.pkl')
```

```python
df_encoded.to_csv(r'C:\Users\BHARGAV\Desktop\insurance-cost-prediction\DataSets\clean_df.csv')
```
